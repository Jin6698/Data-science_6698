```
# read the dataset

import pandas as pd

df = pd.read_csv('/work/wind_dataset.csv')

df.sample(10)
```

```
# find the sum of missing values in each feature
df.isna().sum()
```

```
# drop the missing values

df=df.dropna()
```

```
# prepare the X and y dataframes for the modeling 

X=df.drop(['DATE','WIND'],axis=1)

y=  df['WIND']
```

```
from catboost import CatBoostRegressor

cat = CatBoostRegressor()

cat.fit(X_train,y_train)

y_pred = cat.predict(X_test)

from sklearn.metrics import mean_squared_error

mse_cat = mean_squared_error(y_test,y_pred)

print('mse of the catboost model on the predictions:',mse_cat)

import numpy as np

rmse_cat = np.sqrt(mse_cat)

print('rmse of the catboost model on the predictions:',rmse_cat)

from sklearn.metrics import mean_absolute_error

mae_cat= mean_absolute_error(y_test,y_pred)

print('mae of the catboost model on the predictions',mae_cat)
```

```
from sklearn.ensemble import RandomForestRegressor

rfc = RandomForestRegressor()

rfc.fit(X_train,y_train)

y_pred = rfc.predict(X_test)

from sklearn.metrics import mean_squared_error

mse_rfc = mean_squared_error(y_test,y_pred)

print('mse of the RandomForestRegressor model on the predictions:',mse_rfc)

rmse_rfc = np.sqrt(mse_rfc)

print('rmse of the RandomForestRegressor model on the predictions:',rmse_rfc)

mae_rfc= mean_absolute_error(y_test,y_pred)

print('mae of the RandomForestRegressor model on the predictions',mae_rfc)
```
