```python
import pandas as pd
```

```python

data = pd.read_csv('/content/cancer patient data sets (1).csv')

data.head()
```

```python
'''
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['Level']=le.fit_transform(data['Level'])

data.head()

'''
```

```python
map = {'High': 2, 'Medium': 1, 'Low': 0}
data["Level"].replace(map, inplace=True)

data.head()
```

```python
data['Level'].value_counts()
```

```python
data.isna().sum()
```

```python

X = data.drop(['index','Patient Id','Level'],axis=1)

y=data['Level']
```

```python
X.head()
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE()

# Assuming you have your features in X and labels in y
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)
```

```python
y_smote.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_smote,y_smote,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()

rfc.fit(X_train,y_train)

y_pred = rfc.predict(X_test)

accuracy_score_rfc = accuracy_score(y_test,y_pred)

print(accuracy_score_rfc)
```

```python
from sklearn.naive_bayes import GaussianNB

gb = GaussianNB()

gb.fit(X_train,y_train)

y_pred = gb.predict(X_test)

accuracy_score_gb = accuracy_score(y_test,y_pred)

print(accuracy_score_gb)
```

```python
from sklearn.ensemble import AdaBoostClassifier

Ada = AdaBoostClassifier()

Ada.fit(X_train,y_train)

y_pred = Ada.predict(X_test)

accuracy_score_Ada = accuracy_score(y_test,y_pred)

print(accuracy_score_Ada)
```
